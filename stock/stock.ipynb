                           Stock price prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import yfinance as yf
```

```python
ticker='AAPL'
data=yf.download(ticker,start='2020-01-02',end='2023-01-01')
```

```python
data['Data']=data.index
data['Data']=pd.to_numeric(data['Data'])
```

```python
X=data[['Data']]
y=data['Close']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.index,data['Close'],label='Actual Prices')
plt.plot(X_test.index,prediction,label='Predicted Price',linestyle='--')
plt.xlabel('Data')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
print(f"Model R^2 score: {model.score(X_test,y_test)}")
```
